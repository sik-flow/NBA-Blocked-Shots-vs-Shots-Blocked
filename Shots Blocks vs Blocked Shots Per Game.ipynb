```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib
% matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
blocks = pd.read_csv('blocks.csv')
```

```python
blocks.head()
```

```python
blocks.plot(kind='scatter', x = 'BLK', y = 'BLKA')
```

```python
yogi = blocks.plot(kind='scatter', x = 'BLKA', y = 'BLK', figsize = [11, 11], s = 60, color = 'red', fontsize = 20)
yogi.set_title("Shots Blocked vs Blocked Shots Per Game")
yogi.set_xlabel('Own Shots Blocked', fontsize = 20)
yogi.set_ylabel("Blocked Others' Shots", fontsize = 20)
yogi.text(1.6, 1.35, 'DeMarcus Cousins')
yogi.text(1.1, 1.45, 'Karl-Anthony Towns')
yogi.text(1.02, 1.95, 'Giannis Antetokounmpo')
yogi.text(1.2, 2.4, 'Joel Embiid')
```

```python
yogi = sns.stripplot(x='BLKA', y='BLK', data = blocks, jitter=True, color = 'Red', size = 5) 
yogi.set_xlabel('Times Shot Blocked Per Game')
yogi.text(1.5, 1.5, 'Yogi')
```
